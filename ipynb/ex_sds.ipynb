````python
from obspy.clients.filesystem.sds import Client
from obspy.core.utcdatetime import UTCDateTime
from obspy import read_inventory
client = Client("../SDS")
````

````python
nslc = client.get_all_nslc(sds_type=None, datetime=None)
for n in nslc:
    print(n)
````

````python
stns = client.get_all_stations()
for s in stns:
    print(s)
````

````python
net="BJ"
sta="DHC" 
loc="*" 
cha="BH?" 
stime=UTCDateTime("2018-07-20T16:00:04")
etime=UTCDateTime("2018-08-22T16:00:02")
````

````python
avaper = client.get_availability_percentage(network=net, 
                                            station=sta, 
                                            location=loc, 
                                            channel=cha, 
                                            starttime=stime, 
                                            endtime=etime)
print(avaper)
````

````python
wf = client.get_waveforms(network=net, 
                          station=sta, 
                          location=loc, 
                          channel=cha, 
                          starttime=stime, 
                          endtime=etime)
print(wf.__str__(extended=True))
````

````python
p = wf.plot()
````

````python
invs = read_inventory("../data/BJ.XML")
fig_map = invs.plot(projection="local", resolution="i", label=True, color_per_network=True)
````

````python
inv = invs.select(network="BJ", station="DHC", location="00", channel="BHZ")
fig_resp = inv.plot_response(0.1, output="VEL", show=True)
````
